```python
"""¿Qué sucede si la cantidad de monedas para el cambio es limitada?
Ejm: Si tiene almacenado para el cambio la siguiente cantidad;
1 de 500w, 1 de 100w, 3 de 50w y 2 de 10w.
-¿Cómo se debe distribuir 710 para el cambio?"""

def cambio_moneda2(monedas, cantidad):
    cambios = []
    monedas.sort(reverse=True)

    for valor, cantidad_disponible in monedas:
        cantidad_a_usar = min(cantidad // valor, cantidad_disponible)

        if cantidad_a_usar > 0:
            cambios.append((valor, cantidad_a_usar))
            cantidad -= valor * cantidad_a_usar

        if cantidad == 0:
            break

    if cantidad > 0:
        return "No es posible llegar a el monto con las monedas disponibles."

    return cambios

monedas = [(500, 1), (100, 1), (50, 3), (10, 2)]
cantidad = int(input("Ingrese la cantidad a cambiar: ")) #710
cambios = cambio_moneda2(monedas, cantidad)
print("Monedas: ", monedas)
print("Resultados:", cambios, len(cambios))
```
