```python
def bubble_sort(arr):
    n = len(arr)

    for i in range(n):
        for j in range(0, n-i-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]

    return arr


def selection_sort(arr):
    n = len(arr)

    for i in range(n):
        min_idx = i
        for j in range(i+1, n):
            if arr[j] < arr[min_idx]:
                min_idx = j
        arr[i], arr[min_idx] = arr[min_idx], arr[i]

    return arr


def insertion_sort(arr):
    n = len(arr)

    for i in range(1, n):
        key = arr[i]
        j = i-1
        while j >= 0 and arr[j] > key:
            arr[j+1] = arr[j]
            j -= 1
        arr[j+1] = key

    return arr

arr = [37, 82, 15, 4, 96, 63, 21, 58, 7]
print("Lista original:", arr)

arr = bubble_sort(arr)
print("Lista ordenada (Burbuja):", arr)

arr = selection_sort(arr)
print("Lista ordenada (Selecci贸n):", arr)

arr = insertion_sort(arr)
print("Lista ordenada (Inserci贸n):", arr)

```
