```python
class Deque:
    def __init__(self):
        self.deque = []

    def add_first(self, item):
        self.deque.insert(0, item)

    def remove_first(self):
        if len(self.deque) > 0:
           return self.deque.pop(0)

    def add_last(self, item):
        self.deque.insert(len(self.deque), item)

    def remove_last(self):
        if len(self.deque) > 0:
            return self.deque.pop()

    def esta_vacio(self):
        return len(self.deque) == 0

    def __str__(self):
        return str(self.deque)


deque = Deque()

pre = input("Ingrese elemento: ")
gun = input("Ingrese elemento: ")
ta = input("Ingrese elemento: ")

deque.add_first(pre)
deque.add_last(gun)
deque.add_last(ta)

print()
print("elementos:", deque)
print()

primer_elemento = deque.remove_first()
segundo_elemento = deque.remove_last()

print(f"Primer elemento removido: {primer_elemento}")
print(f"Ultimo elemento removido: {segundo_elemento}")
print()

print(f"Deque actual (primer y último elementos removidos): {deque} ")
lista
```
